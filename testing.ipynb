```python
import numpy as np
import matplotlib.pyplot as plt

radius = 8
i, j = radius - 1, radius - 1
matrix = np.zeros((radius*2 - 1, radius*2 - 1))

i_l_outside = []
j_l_outside = []
i_l_inside = []
j_l_inside = []

for i_p in range(i - radius, i+radius):
    for j_p in range(j- radius, j + radius):
        if (i_p - i) ** 2 + (j_p - j) ** 2 < radius ** 2:
            if ((i_p < i - 1) or (i_p > i + 1)) or ((j_p < i - 1) or (j_p > i + 1)):
                # inside += 1
                # matrix[i_p, j_p] = 1
                i_l_outside.append(i - i_p)
                j_l_outside.append(j - j_p)
            else:
                # outside += 1
                # matrix[i_p, j_p] = 0
                i_l_inside.append(i - i_p)
                j_l_inside.append(j - j_p)
                
# Plot the matrix with a filled circle
plt.imshow(matrix, cmap='binary', interpolation='none')
plt.show()
matrix_2 = np.zeros((30, 30))
i, j = 15, 15
for el in range(len(i_l_outside)):
    matrix_2[i - i_l_outside[el]][j - j_l_outside[el]] = 1
for el in range(len(i_l_inside)):
    matrix_2[i - i_l_inside[el]][j - j_l_inside[el]] = 0.5
plt.imshow(matrix_2, cmap='Blues', interpolation='none')
plt.show()
```
