### Zadanie 2

```python
x = np.linspace(0, 2*np.pi)

f_x = np.sin(x)
g_x = np.cos(x) ** 2
h_x = np.cos(x) ** 3

plt.figure(figsize=(10,5))
plt.plot(x, f_x, color='red', linestyle='dotted', label='sin(x)')
plt.plot(x, g_x, color='blue', label='cos^2(x)')
plt.plot(x, h_x, color='green', linestyle='dashed', label='cos^3(x)')

plt.title('Wykres trzech funkcji')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.legend()
plt.grid()
plt.show()
```
