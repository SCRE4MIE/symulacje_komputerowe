```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
def dyskretny_model_wzrostu_logistycznego(r, x_0, n):
    result = []
    x_t = x_0
    for i in range(n):
        x_t = r*x_t*(1-x_t)
        result.append(x_t)
    return result
```

## Zadanie 1 a)

```python
# t = 0,...100
# r = 1.1
# x_0 nale≈ºy do (0,1)

tab_x_0 = np.random.rand(5,1)
t = [t for t in range(100)]
r = 1.1
```

```python
plt.figure(figsize=(15,15))
colors = ['red', 'blue', 'yellow', 'green', 'purple']
for i in range(4):
    Y = dyskretny_model_wzrostu_logistycznego(r=r, x_0=tab_x_0[i][0], n=len(t))
    plt.scatter(x=t, y=Y, color=colors[i], label=f'x_0={tab_x_0[i][0]}', alpha=0.7)
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```

## Zadanie 1 b)

```python
def sprawdz_dla_ktorego_t(r, x_0, w, c, n):
    x_t = x_0
    for t in range(n):
        x_t = r*x_t*(1-x_t)
        if abs( x_t - w ) < c:
            return t
    return None
```

## Zadanie 1 e)

```python
def dyskretny_model_wzrostu_logistycznego_odciecie(r, x_0, n, cut):
    result = []
    x_t = x_0
    for i in range(n):
        x_t = r*x_t*(1-x_t)
        result.append(x_t)
    return result[cut[0]:cut[1]]
```

```python
# r nale≈ºy do [2.8, 4]
# x_0 = 0.5

x_0 = 0.5
t = [t for t in range(300)]
R = np.linspace(2.8, 4, num=300)

plt.figure(figsize=(15,15))
for r in R:
    Y = dyskretny_model_wzrostu_logistycznego_odciecie(r=r, x_0=x_0, n=len(t), cut=[0, 300])
    plt.scatter(x=[r for i in range(len(Y))], y=Y, color='blue', alpha=0.7, s=5)
plt.xlabel('r')
plt.ylabel('x')
plt.title('Przed odrzuceniem pierwszych 249 warto≈õci')
plt.show()
```
